```python
import os
import datetime
import logging
from groq import Groq
from dotenv import load_dotenv
```

```python
load_dotenv()

client = Groq(api_key=os.getenv("api_key"))

def LLM(prompt):
    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[{"role": "user", "content": prompt}],
        max_tokens=150,
        temperature=0.7
    )
    return response.choices[0].message.content
```

```python
logging.basicConfig(
    filename="pipeline.log",
    level=logging.INFO,
    format="%(asctime)s - %(message)s"
)
```

```python
problem = """
A small business wants to increase online sales but has limited budget.
What strategy should they follow?
"""

cot_prompt = f"""
You are a business expert.

<instruction>
Break the problem into 3-4 clear sub-questions.
Do NOT give vague or generic questions.
</instruction>

<problem>
{problem}
</problem>
"""

sub_questions = LLM(cot_prompt)
```

```python
solve_prompt = f"""
<instruction>
Answer each question clearly and practically.
Do NOT give vague or unnecessary explanations.
</instruction>

<questions>
{sub_questions}
</questions>
"""

answers = LLM(solve_prompt)
```

```python
final_prompt = f"""
<instruction>
Give a final actionable recommendation.
Do NOT repeat previous answers.
Keep it concise and practical.
</instruction>

<answers>
{answers}
</answers>
"""

final_answer = LLM(final_prompt)
```

```python
logging.info("New Run")
logging.info(f"Problem:\n{problem}")
logging.info(f"Step 1 - Sub Questions:\n{sub_questions}")
logging.info(f"Step 2 - Answers:\n{answers}")
logging.info(f"Step 3 - Final:\n{final_answer}")

print("Done")
```
