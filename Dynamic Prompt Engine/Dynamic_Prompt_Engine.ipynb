```python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
load_dotenv()
```

```python
topic = input("Enter topic: ")
level = input("Enter audience level (beginner/intermediate/expert): ")

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful teacher."),
    ("human", "Explain {topic} for a {level} level audience in simple terms.")
])

formatted_prompt = prompt.format_messages(topic=topic, level=level)
```

```python
groq_model = ChatOpenAI(
    model="llama-3.3-70b-versatile",
    temperature=0.7,
    openai_api_key=os.getenv("GROQ_API_KEY"),
    openai_api_base="https://api.groq.com/openai/v1"
)

groq_model_2 = ChatOpenAI(
    model="openai/gpt-oss-120b",
    openai_api_key=os.getenv("GROQ_API_KEY"),
    openai_api_base="https://api.groq.com/openai/v1"
)
```

```python
def stream_output(model, name):
    print(f"\n {name} Response\n")
    
    for chunk in model.stream(formatted_prompt):
        print(chunk.content, end="", flush=True)
    
    print("\n")
```

```python
stream_output(groq_model, "Groq llama 70B versatile")
stream_output(groq_model_2, "OpenAI GPT oss 120B")
```
